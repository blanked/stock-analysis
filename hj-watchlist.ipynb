```python
import yfinance as yf
import json
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import string
```

### Download market data
First we need to download all the stock tickers. 
TODO: Identify what are all the tickers

```python
allTickers = []

for alphabet in string.ascii_lowercase:
    page = requests.get(f"http://eoddata.com/stocklist/NASDAQ/{alphabet}.htm")
    contents = page.content
    soup = BeautifulSoup(contents, 'html.parser')
    rows = soup.find("div", {"id": "ctl00_cph1_divSymbols"}).find_all("tr", {"class": ["ro", "re"]})
    tickers = [row.find("a").text for row in rows]
    allTickers.extend(tickers)
```

```python
len(allTickers)
```

```python
print(allTickers)
```

```python
tickers = {}
for ticker in allTickers:
    tickers[ticker] = yf.Ticker(ticker)
```
