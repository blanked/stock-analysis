### Adding in change data

```python
newColumns = []
formats = {}
for column in financials.columns:
    newColumn = column + " change"
    financials[newColumn] = (financials[column] - financials[column].shift(-1)).divide(financials[column].shift(-1))
    formats[newColumn] = '{:,.2%}'.format
    financials[newColumn] = financials[newColumn].map('{:,.2%}'.format)
    
financials
```
